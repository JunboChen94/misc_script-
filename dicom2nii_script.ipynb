```python
import os 
```

```python
SAVE_PATH = "./"
stdouts = []
i = 0
for folder in os.listdir(SAVE_PATH)[:20]:
    for subfolder in os.listdir(os.path.join(SAVE_PATH, folder)):
        name = os.path.join(SAVE_PATH, folder, subfolder)
        if not os.path.isdir(name):
            continue
        subfolder2 = [x for x in os.listdir(name) if os.path.isdir(os.path.join(name, x))]
        for n in subfolder2:
            stdout = os.popen('dcm2nii -o {} {}'.format(name, \
                                                        os.path.join(name, n))).read()
            stdouts.append(stdout)
        i += 1
        print("{}th processed".format(i))
```
