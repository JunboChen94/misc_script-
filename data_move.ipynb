```python
import os
import subprocess
```

Define PARTH

```python
SOURCE_PATH = './CARE_SUBSET/update_data/DTI_Final_4bof4_092319_310/'
SAVE_PATH = './CARE_SUBSET/nifti/DTI_Final_4bof4'
if not os.path.isdir(SAVE_PATH):
    os.mkdir(SAVE_PATH)
```

```python
def find_nth(haystack, needle, n):
    start = haystack.find(needle)
    while n > 1 and start > 0:
        start = haystack.index(needle, start+len(needle))
        n -= 1
    return start         
```

Unzip and Manage subject files

```python
filenames = sorted(os.listdir(SOURCE_PATH))
for i, filename in enumerate(filenames): 
    if ('.zip' not in filename) and ('_bvals' not in filename) and ('_bvecs' not in filename):
        continue
    subjectkey = filename[:find_nth(filename, '_', 3)]
    subjectkey_folder_name = os.path.join(SAVE_PATH, subjectkey)
    if not os.path.isdir(subjectkey_folder_name):
        os.mkdir(subjectkey_folder_name)
    if '.zip' in filename:
        zipkey = filename.replace('.zip', '')
        dicom_folder_name = os.path.join(subjectkey_folder_name, zipkey)
        if not os.path.isdir(dicom_folder_name):
            os.mkdir(dicom_folder_name)
        os.system('unzip {} -d {}'.format(os.path.join(SOURCE_PATH, filename), \
                                          dicom_folder_name))
    else:
        os.system('cp {} {}'.format(os.path.join(SOURCE_PATH, filename),\
                                    os.path.join(subjectkey_folder_name, filename)))
    
    if (i+1) % 100 == 0:
        print("{}/{}".format(i+1, len(os.listdir(SOURCE_PATH))))
```

Default dicom files directory is too long, move it to shorter directory 

```python
for folder in os.listdir(SAVE_PATH):
    name = os.path.join(SAVE_PATH, folder)
    for folder2 in os.listdir(name):
        zipfolder = os.path.join(name, folder2)
        if not os.path.isdir(zipfolder):
            continue
        subnames = list(os.walk(zipfolder))
        files_folder = [x[0] for x in subnames if "DICOM/files" in x[0]]
        if len(files_folder) != 1:
            print("{} has no or more than 2 DICOM files folder".format(files_folder))
        else:
            dicome_new_folder = os.path.join(name, "DICOMs_{}".format(folder2[-5:]))
            if not os.path.isdir(dicome_new_folder):
                os.mkdir(dicome_new_folder)
            os.system("mv {}/* {}".format(files_folder[0], dicome_new_folder))
```
